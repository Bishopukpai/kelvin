## Dataset Description

| Column_Name | Value | Description |
| --- | --- | --- |
| age | number | patient's age (in years) |
| sex | 1 or 0 | patient's gender (1: male , 0:female) |
| cp | 0,1,2 or 3 | Chest pain type: 0: asymptomatic 1: typical angina 2: atypical angina 3: non-anginal pain |
| trestbps | number | Resting blood pressure (in mm Hg) |
| chol | number | Serum cholestoral in mg/dl |
| fbs | 1 or 0 | if the patient's fasting blood sugar > 120 mg/dl (1: yes; 0: no) |
| restecg | 1 or 0 | Resting ECG results: 0: normal 1: ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) 2: probable or definite left ventricular hypertrophy by Estes' criteria |
| thalach | number | Maximum heart rate achieved. |
| exang | 1 or 0 | Exercise induced angina (1: yes; 0: no) |
| oldpeak | number | ST depression induced by exercise relative to rest.|
| slope | 0, 1 or 2 | The slope of the peak exercise ST segment: 0: upsloping 1: flat 2: downsloping |
| ca | number | Number of major vessels (0â€“3) colored by flourosopy|
| thal | number | Thalassemia types|
| target | 1 or 0 | Whether the patient has hypertension (1) or not (0) |

## Model building

```python
# split data into dependent and independent features
x = df_hyp.drop('target', axis=1)
y = df_hyp['target']
```

```python
# splitting dataset int training and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# storing all algorithm for use in a dictionary for easy integration
models={
    "Logisitic Regression" :LogisticRegression(max_iter=20000),
    "Decision Tree" :DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Multinomial Naive Bayes": MultinomialNB(),
    "Support Vector Machine" : SVC()
}
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train,y_train) # Train Model
    # Make predictions
    y_train_pred = model.predict(x_train)
    y_test_pred =  model.predict(x_test)

    print(list(models.keys())[i])
    
    # Training set performance
    print('Model performance for Training set ')
    print(f'- Accuracy: {accuracy_score(y_train, y_train_pred):.4f}')
    print(f'- F1 score: {f1_score(y_train, y_train_pred):.4f}')
    print(f'- Precision: {precision_score(y_train, y_train_pred ):.4f}')
    print(f'- Recall: {recall_score(y_train, y_train_pred):.4f}')

    print('-'*35)
    
     # Test set performance
    print('Model performance for Test set ')
    print(f'- Accuracy: {accuracy_score(y_test, y_test_pred):.4f}')
    print(f'- F1 score: {f1_score(y_test, y_test_pred):.4f}')
    print(f'- Precision: {precision_score(y_test, y_test_pred ):.4f}')
    print(f'- Recall: {recall_score(y_test, y_test_pred):.4f}')


    print(f'*'*35)
    print('\n')
```

```python
# we will use Random Forest algorithm to build our model 
# since it's one of the model that performs more better

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
print(f'Fl score: {f1_score(y_test, y_pred):.4f}')
```
