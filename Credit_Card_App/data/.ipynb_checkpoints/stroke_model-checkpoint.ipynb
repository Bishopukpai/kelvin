```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
df_str = pd.read_csv("stroke_data.csv")

df_str.head()
```

```python
df_str.info()
```

## Data cleaning

```python
df_str.isnull().sum()
```

```python
# removig missing observations
df_str.dropna(inplace=True)
```

```python
df_str['sex'] = df_str['sex'].astype(np.int64)
df_str['age'] = df_str['age'].astype(np.int64)
```

```python
df_str.info()
```

```python
str_viz = df_str.copy()
```

```python
str_viz['sex'] = str_viz['sex'].map({1 : "Male", 0: "Female"}) 

columns = ['hypertension', 'heart_disease', 'ever_married', 'smoking_status', 'stroke']

for column in columns:
    str_viz[column] = str_viz[column].map({1 : "Yes", 0: "No"})

str_viz['work_type'] = str_viz['work_type'].map({0 : "Never_worked", 1: "children",
                                                2: "Govt_job", 3: "Self-employed", 4: "Private"})

str_viz['Residence_type'] = str_viz['Residence_type'].map({1 : "Urban", 0: "Rural"})
```

```python
# function for bar plot visualization
def bar(column, title, xlabel, ylabel):
    str_viz[column].value_counts().plot(kind='bar');
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
plt.figure(figsize=[8,5])
bar("stroke", "cases of patients with stroke", "stroke", "count")
```

```python
# split data into dependent and independent features
x = df_str.drop('stroke', axis=1)
y = df_str['stroke']
```

```python
# splitting dataset int training and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# storing all algorithm for use in a dictionary for easy integration
models={
    "Logisitic Regression" :LogisticRegression(max_iter=20000),
    "Decision Tree" :DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine" : SVC()
}
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train,y_train) # Train Model
    # Make predictions
    y_train_pred = model.predict(x_train)
    y_test_pred =  model.predict(x_test)

    print(list(models.keys())[i])
    
    # Training set performance
    print('Model performance for Training set ')
    print(f'- Accuracy: {accuracy_score(y_train, y_train_pred):.4f}')
    print(f'- F1 score: {f1_score(y_train, y_train_pred):.4f}')
    print(f'- Precision: {precision_score(y_train, y_train_pred ):.4f}')
    print(f'- Recall: {recall_score(y_train, y_train_pred):.4f}')

    print('-'*35)
    
     # Test set performance
    print('Model performance for Test set ')
    print(f'- Accuracy: {accuracy_score(y_test, y_test_pred):.4f}')
    print(f'- F1 score: {f1_score(y_test, y_test_pred):.4f}')
    print(f'- Precision: {precision_score(y_test, y_test_pred ):.4f}')
    print(f'- Recall: {recall_score(y_test, y_test_pred):.4f}')


    print(f'*'*35)
    print('\n')
```

```python
# we will use Random Forest algorithm to build our model 
# since it's one of the model that performs more better

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
print(f'Fl score: {f1_score(y_test, y_pred):.4f}')
```
